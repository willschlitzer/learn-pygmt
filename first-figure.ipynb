```python
import pygmt
```

```python
fig=pygmt.Figure()
fig.coast(region=[-69, -68, 43.75, 44.75], shorelines=True)
fig.show()
```

```python
fig=pygmt.Figure()
fig.coast(region=[-69, -68, 43.75, 44.75], shorelines=True, land="lightgreen", water="lightblue")
fig.show()
```

```python
fig=pygmt.Figure()
fig.coast(region=[-69, -68, 43.75, 44.75], shorelines=True, land="lightgreen", water="lightblue", projection="M10c")
fig.show()
```

```python
fig=pygmt.Figure()
fig.coast(region=[-69, -68, 43.75, 44.75], shorelines=True, land="lightgreen", water="lightblue", projection="M10c", frame=True)
fig.show()
```

```python
fig=pygmt.Figure()
fig.coast(region="d", shorelines=True, land="lightgreen", water="lightblue", projection="W12c", frame=True)
fig.show()
```

```python
fig=pygmt.Figure()
fig.coast(region=[-69, -68, 43.75, 44.75], shorelines=True, land="lightgreen", water="lightblue", projection="M10c", frame=["ag", '+t"Coastal Maine"'])
fig.show()
```

```python
fig = pygmt.Figure()
fig.coast(
    region=[-69, -68, 43.75, 44.75],
    shorelines=True,
    land="lightgreen",
    water="lightblue",
    projection="M10c",
    frame=["a", '+t"Coastal Maine"'],
)
fig.show()
```
