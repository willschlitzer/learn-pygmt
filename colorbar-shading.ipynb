```python
import pygmt
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=[-86, -64 , 17, 24])
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="M6i", frame=True, cmap="geo")
fig.colorbar(frame=["x+lElevation", "y+lm"])
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="M6i", frame=True, cmap="geo")
fig.colorbar(frame=["x+lElevation", "y+lm"], I=True)
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="M6i", frame=True, cmap="geo")
fig.colorbar(frame=["x+lElevation", "y+lm"], I=False)
fig.show()
```
