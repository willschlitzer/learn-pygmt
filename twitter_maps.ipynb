```python
import pygmt
```

```python
grid = pygmt.datasets.load_earth_relief(resolution="01s", region=[127.5, 128.50, 26, 27])
fig = pygmt.Figure()
#fig.grdimage(region="JP", grid=grid,cmap="geo", frame=True)
fig.grdimage(grid=grid,cmap="geo", frame=True)
fig.savefig("oki.png")
fig.show()
```
