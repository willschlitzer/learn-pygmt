```python
import pygmt
```

```python
cmap="ex_cpt.cpt"
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=[-108, -103, 35, 40])
```

```python
#pygmt.makecpt(cmap="bathy", series="1000/4000/1000", continuous=True, output=cmap)
cmap="haxby"
```

```python
fig = pygmt.Figure()
projection="M4i"
fig.basemap(perspective=[135, 30], frame=["xa", "yaf", "z2+1km"], projection=projection)
fig.grdview(grid=grid, perspective=[135, 30], plane="1000+gblue", surftype="s", projection=projection, cmap=cmap, zsize="0.5i", contourpen="0.1p")
fig.colorbar(perspective=[225, 30], cmap=cmap)

fig.show()
```

```python
fig = pygmt.Figure()
#gradient = fig.grdgradient(grid=grid)
#gradient = pygmt.grdgradient(grid=grid)
fig.grdimage(grid=grid, cmap=cmap, projection="M4i",frame=True)
fig.grdcontour(annotation=1000, interval=200, grid=grid, projection="M4i",frame=True)
fig.show()
```
