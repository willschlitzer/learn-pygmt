```python
import pygmt
```

```python
# Hammer projection with default central meridian
fig = pygmt.Figure()
# Use region "d" to specify global region (-180/180)
fig.coast(region="d", projection="H12c", land="black", water="cornsilk", frame="afg")
fig.show()
```
