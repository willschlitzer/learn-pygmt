```python
import pygmt
from pygmt.helpers.testing import load_static_earth_relief
```

```python
grid = load_static_earth_relief()
```

```python
print(pygmt.grdinfo(grid=grid))
```

```python
fig = pygmt.Figure()
fig.grdcontour(grid=grid, interval=50, annotation=200, projection="M10c", frame=True)
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdcontour(grid=grid,         
        interval="10",
        annotation="50",
        projection="M10c",
        pen=["a1p,red", "c0.5p,black"],
        label_placement="d6c",
        frame=True,)
fig.show()
```

```python
grid_ = grid.sel(lat=slice(-20, -10))

fig = pygmt.Figure()
fig.grdcontour(grid=grid_, interval="100", projection="M10c", frame=True)
fig.show()
```

```python
fig = pygmt.Figure()

comargs = {
        "region": [-53, -49, -20, -17],
        "grid": grid,
        "interval":50,
        "annotation":100,
        "resample": "100",
        "projection": "M10c",
        "cut": 10,
    }
fig.grdcontour(**comargs, limit=(0, 649), pen=["a1p,blue", "c0.5p,blue"], frame=True)
fig.grdcontour(**comargs, limit=(650, 1000), pen=["a1p,black", "c0.5p,black"])
fig.show()
```
