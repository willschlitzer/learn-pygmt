```python
import pygmt
from pygmt.helpers.testing import load_static_earth_relief
grid = load_static_earth_relief()
region = [-55, -50, -18, -12]
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid)
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, projection="Q15c+", perspective=[135, 15], frame=True)
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, perspective=[225, 30], zscale=0.005)
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, perspective=[225, 30], zsize="10c")
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, cmap="oleron", surftype="i")
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, cmap="oleron", surftype="s+m")
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, plane=100, perspective=[225, 30], zscale=0.005)
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, plane="100+ggray", perspective=[225, 30], zscale=0.005)
fig.show()
```

```python
fig = pygmt.Figure()
projection = f"T{(region[0]+region[1])/2}/{abs((region[2]+region[3])/2)}"
fig.grdview(grid=grid,
        projection=projection,
        perspective=[225, 30],
        zscale=0.005,
        frame=["xaf", "yaf", "zaf"],
    )
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, cmap="relief", surftype="s", contourpen="0.5p,black,dash")
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid, cmap="relief", surftype="sm", meshpen="0.5p,black,dash")
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdview(grid=grid,
        plane=100,
        perspective=[225, 30],
        zscale=0.005,
        facadepen="0.5p,blue,dash",
    )
fig.show()
```

```python
drapegrid = 1.1 * grid
fig = pygmt.Figure()
fig.grdview(grid=grid, drapegrid=drapegrid, cmap="oleron", surftype="c")
fig.show()
```
