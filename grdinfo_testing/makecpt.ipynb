```python
import pygmt
from pygmt.helpers.testing import load_static_earth_relief
grid = load_static_earth_relief()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="oleron", series=[0, 1000])
fig.grdimage(grid=grid, projection="M10c", frame=True)
fig.show()
```

```python
print(pygmt.grdinfo(grid=grid))
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="relief")
fig.grdimage(grid=grid, projection="M10C", frame=True)
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="rainbow", truncate=[0.15, 0.85], series=[0, 1000])
fig.grdimage(grid=grid, projection="M10c", frame=True)
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="earth", reverse=True, series=[0, 1000])
fig.grdimage(grid=grid, projection="M10c", frame=True)
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="earth", reverse="cz", series=[0, 1000])
fig.grdimage(grid=grid, projection="M10c", frame=True)
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="blue,white", continuous=True, series=[0, 1000])
fig.grdimage(grid=grid, projection="M10c", frame=True)
fig.show()
```
