```python
import pygmt
```

```python
#Transverse Mercator
fig = pygmt.Figure()
fig.coast(region=[20, 50, 30, 45], projection="T35/12c", land="lightbrown", water="seashell", shorelines="thinnest", frame="afg")
fig.show()
```

```python
#Universal Transverse Mercator
fig = pygmt.Figure()
fig.coast(region=[127.5, 128.5, 26, 27], projection="U52R/12c", land="lightgreen", water="lightblue", shorelines="thinnest", frame="afg")
fig.show()
```

```python
#Oblique Mercator
fig = pygmt.Figure()
fig.coast(region=[270, 305, 20, 25], projection="M12c", land="lightgreen", water="lightblue", shorelines="thinnest", frame="afg")
fig.show()
```
