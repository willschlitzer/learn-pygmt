```python
import pygmt
```

```python
fig = pygmt.Figure()
```

```python
fig_Kwargs = dict(grid="@earth_relief_01s",
              region=[-92.25,-91.75, 1.5, 1.9],
              projection="M4i",
              frame=True)
```

```python
fig.grdimage(cmap="haxby", **fig_Kwargs)
```

```python
fig.grdcontour(annotation=500, interval=100, **fig_Kwargs)
```

```python
fig.colorbar()
```

```python
fig.show()
```
