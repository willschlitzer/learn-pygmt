```python
import pygmt
```

```python
fig = pygmt.Figure()
```

```python
fig_Kwargs = dict(grid="@earth_relief_15s",
              region=[-92,-90, -1.5, .5],
              projection="M4i",
              frame=True)
```

```python
fig.grdimage(cmap="haxby", **fig_Kwargs)
```

```python
fig.grdcontour(annotation=1000, interval=250, **fig_Kwargs)
```

```python
fig.colorbar()
```

```python
fig.show()
```
