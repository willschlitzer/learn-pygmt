Creating a map with contour lines
=================================

Using a grid file to create a map with contour lines is handled by the "grdcontour" method of the class pygmt.Figure

```python
import pygmt
```

## Create contour plot

The grdcontour method takes the grid input and the region region values. It plots annotated contour lines, which are darker and have the elevation/depth written on them, and unannotated contour lines. In the example blow, the default contour line intervals are 500 meters, with an annotated contour line every 1000 meters. By default, it plots the map with a Mercator projection and with no frame.

```python
fig = pygmt.Figure()
fig.grdcontour(grid="@earth_relief_15s",region=[-92,-90, -1.5, .5])
fig.show()
```

## Contour line settings

To change the settings for the contour lines, the parameters of "annotation" and "interval" are used. Interval sets the vertical distance between contour lines, and annotation sets the interval for annotated contour lines. In the example below, there are contour intervals every 250 meters and an annotated contour line every 1,000 meters.

```python
fig = pygmt.Figure()
fig.grdcontour(annotation=1000, interval=250, grid="@earth_relief_15s",
              region=[-92,-90, -1.5, .5])
fig.show()
```

## Contour limits

The limit parameter sets the minimum and maximum values for the contour lines, with the value. The paraemeter accepts the low and high values, and is either a list (as below) or a string  limit="-4000/0"

```python
fig = pygmt.Figure()
fig.grdcontour(annotation=1000, interval=250, grid="@earth_relief_15s",
              region=[-92,-90, -1.5, .5], limit=[-4000,0])
fig.show()
```

## Map settings

The grdcontour method accepts additional parameters, including setting the projection and frame.

```python
fig = pygmt.Figure()
fig.grdcontour(annotation=1000, interval=250, grid="@earth_relief_15s",
              region=[-92,-90, -1.5, .5], limit=[-4000, 0], projection="M4i",frame=True)
fig.show()
```
