```python
import pygmt
```

```python
pygmt.makecpt(cmap="haxby", series="-3500/500/1000", continuous=True, output="galcpt.cpt")
ga_fig = pygmt.Figure()
#ga_fig.grdimage(shading=True, **Ga_KWARGS)
ga_fig.grdimage(cmap="galcpt.cpt",
                grid="@earth_relief_01m",
                region=[-92.5,-89, -1.5, 2],
                projection="M4i",
                frame=True)
ga_fig.grdcontour(grid="@earth_relief_01m",
                  region=[-92.5,-89, -1.5, 2],
                  annotation=1000,
                  interval=500,
                  projection="M4i",
                  frame=True)
ga_fig.colorbar(cmap="galcpt.cpt")
#ga_fig.show()
```
