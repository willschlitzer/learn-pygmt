```python
import pygmt
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=[128.5, 145, 30, 46])
```

```python
pygmt.grdinfo(grid=grid)
```

```python
#pygmt.makecpt(cmap="bathy", series="-10000/0/2000", continuous=True, output="galcpt.cpt")
fig = pygmt.Figure()
fig.grdimage(#cmap="galcpt.cpt",
                grid=grid,
                projection="M4i",
                frame=True)
fig.colorbar(cmap="galcpt.cpt")
fig.show()
```
