```python
import pygmt
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=[-86, -64 , 17, 24])
```

```python
pygmt.grdinfo(grid=grid)
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid)
#fig.colorbar(frame=["x+lElevation", "y+lm"])
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid, cmap="geo")
#fig.colorbar(frame=["x+lElevation", "y+lm"])
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid, frame=True, projection="M6i")
fig.show()
```

```python
fig = pygmt.Figure()

fig.grdimage(grid=grid, frame=True, projection="M6i")
fig.colorbar(frame=["x+lElevation", "y+lm"])
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="M6i", frame=True, cmap="geo")
fig.colorbar(frame=["x+lElevation", "y+lm"])
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdimage(grid=grid, projection="M6i", frame=True, cmap="geo")
fig.colorbar(frame=["x+lElevation", "y+lm"])
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="geo", series="-8500/2500/1000", continuous=True)
fig.grdimage(grid=grid, projection="M6i", frame=True)
fig.colorbar()
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="geo", series=[-8500,2500,1000], continuous=True)
fig.grdimage(grid=grid, projection="M6i", frame=True)
fig.colorbar()
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="geo", series="-8000/2500/1000", truncate="-8000/2500", continuous=True)
fig.grdimage(grid=grid, projection="M6i", frame=True)
fig.colorbar()
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(series="-10000/10000/500", continuous=True)
fig.grdimage(grid=grid, projection="M6i", frame=True)
fig.colorbar()
fig.show()
```

```python
fig = pygmt.Figure()
pygmt.makecpt(cmap="geo", series="-8500/2500/2000", continuous=True)
fig.grdimage(grid=grid, projection="M6i", frame=True)
fig.colorbar()
fig.show()
```
