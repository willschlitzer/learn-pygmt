```python
import pygmt
from pygmt import Figure, grd2cpt
grid = pygmt.datasets.load_earth_relief(resolution="01d")
#region = 
```

```python
fig = pygmt.Figure()
#fig.basemap(region=[0,10,0,10], projection="X15c", frame="a")
#pygmt.grd2cpt(grid=grid) # works
pygmt.grd2cpt(grid=grid,
              cmap="rainbow",#) #works
              #truncate=[0.15, 0.85],#) #works
              continuous=True,) #works
              #series=[-4500, 1500, 500], #) #works
              #reverse=True,#) #works
              #region=[-85, -66, 18, 23]) # works
fig.grdimage(grid, projection="W0/15c")
fig.colorbar(frame="a2000")
fig.show()
```

```python
print(pygmt.grdinfo(grid=grid))
```

```python
fig_ref = pygmt.Figure()

pygmt.grd2cpt(grid=grid, cmap="rainbow", continuous=True,)
fig_ref.grdimage(grid, projection="W0/15c")
fig_ref.colorbar(frame="a2000")
fig_ref.show()
```
