```python
fig = pygmt.Figure()
fig.basemap(projection="X10c", region=[0,10,0,10], frame=True)
fig.plot(x=[1,4], y=[2,3], pen="black,-")
fig.plot(x=[2,4], y=[2,3], pen="dot")
fig.show()
```
