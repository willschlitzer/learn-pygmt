```python
import pygmt
import pandas as pd
from pygmt.src import which
```

```python
hotspots = pygmt.datasets.load_hotspots()
print(hotspots)
```

```python
fname = pygmt.src.which("@hotspots.txt", download="c")
```

```python
data = pd.read_csv(fname, header=3, sep=r"\s+", names=["longitutde", "latitude", "icon_size", "name"])
```

```python
with open(hotspots) as f:
    f.readline()
    f.readline()
    f.readline()
    hotspots = []
    for line in f:
        line_split = line.strip().split("\t")
        hotspot = [float(item.strip()) for item in line_split[0].split()]
        hotspot.append(line_split[1].title())
        hotspots.append(hotspot)
print(hotspots)
```

```python
pd.DataFrame(hotspots, columns=["longitude", "latitude", "icon_size", "name"])
```
