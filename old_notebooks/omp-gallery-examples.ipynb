```python
import pygmt
```

```python
fig = pygmt.Figure()
# Using the origin and azimuth
fig.coast(projection="Oa-120/25/-30/12c+v",region= "-122/35/-107/22+r",frame="afg",
          shorelines="1/thin", land="gray",water="lightblue",
)
fig.show()
```
