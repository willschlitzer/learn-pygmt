```python
import pygmt
```

```python
fig = pygmt.Figure()
cpt = pygmt.makecpt(cmap="no_green", series=[-2, 30, 2])
fig.grdimage(region="g", projection="W180/15c", cmap=cpt, frame=True, grid="@otemp.anal1deg.nc?otemp[2,0]")
fig.show()
```
