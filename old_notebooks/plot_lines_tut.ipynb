```python
import pygmt
from numpy import random
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c/10c", frame="af")
fig.plot(x=[3, 5, 6], y=[4, 8, 4], )#pen="3p,red")
fig.plot(x=[2, 4, 3], y=[4, 8, 4], pen="3p,red")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c/10c", frame="af")
x1 = random.randint(10, size=5)
y1 = random.randint(10, size=5)
fig.plot(x=x1, y=y1, )#pen="3p,red")
fig.plot(x=[2, 4, 3], y=[4, 8, 4], pen="3p,red")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c/10c", frame="af")
x1 = random.randint(10, size=5)
y1 = random.randint(10, size=5)
fig.plot(x=x1, y=y1, cmap="geo", pen="3p")
fig.show()
```
