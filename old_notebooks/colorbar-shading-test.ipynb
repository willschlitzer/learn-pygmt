```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c", frame="af")
fig.colorbar(cmap="geo")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c", frame="af")
fig.colorbar(cmap="geo", shading=True)
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c", frame="af")
fig.colorbar(cmap="geo", I=True)
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c", frame="af")
fig.colorbar(cmap="geo", shading="0.5")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X15c", frame="af")
fig.colorbar(cmap="geo", shading=[-0.7, 0.2])
fig.show()
```
