```python
import pygmt
import pandas as pd
import datetime
```

```python
fig = pygmt.Figure()
fig.coast(R="d", J="W0/15c", B="a", land="lightgray")
fig.solar(R="d", J="W0/15c", B="a", T="d+d1990-02-17T04:55:00")
fig.show()
```

```python
fig = pygmt.Figure()
fig.coast(R="d", J="W0/15c", B="a", land="lightgray")
fig.solar(R="d", J="W0/15c", B="a", terminator="invalid", terminator_datetime="1990-02-17 04:55:00")
fig.show()
```

```python
time = datetime.datetime(year=1990, month=2, day=17,  hour=4, minute=55)
fig = pygmt.Figure()
fig.coast(R="d", J="W0/15c", B="a", land="lightgray")
fig.solar(R="d", J="W0/15c", B="a", terminator="d", terminator_datetime=time)
fig.show()
```

```python
fig = pygmt.Figure()
fig.coast(R="d", J="W0/15c", B="a", land="lightgray")
fig.solar(R="d", J="W0/15c", B="a", terminator="d", terminator_datetime="1990-02-17 04:25:00")
fig.show()
```
