```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.coast(shorelines="1/0.5p", region=[-88, -70, 18, 24], projection="D-79/21/19/23/12c", land="lightgreen", water="lightblue", frame="afg")
fig.show()
```

```python
fig = pygmt.Figure()
fig.coast(
    shorelines="1/0.5p",
    region=[-130, -70, 24, 52],
    projection="D-100/35/33/45/12c",
    land="gray",
    borders=["1/thick,black", "2/thin,black"],
    frame="afg",
)

fig.show()
```
