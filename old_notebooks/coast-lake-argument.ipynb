```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.coast(
    #region="-30/30/-40/40", #R
    region="US.UT", #R
    projection="M25c", #J
    frame="afg", #B
    rivers="1/1p,black", #I
    borders="1/0.5p,-", #N
    shorelines="0.25p,white", #W
    land="moccasin", #G
    water="skyblue", #S
    resolution="i", #D
    area_thresh=1000, #A
    #map_scale="jMM+c1+w1000k+f+l", #L
    timestamp=True, #U
    verbose="q", #V
    xshift = "a4c", #X
    yshift = "a10c", #Y
    transparency = 13, #t
    dcw="MA+gred",
    #C=["blue+l", "red+r"]
    C="red+r"
)
fig.show()
```
