```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.plot(region=[0, 10, 0, 10], projection="X25c/20c", frame="a", x=[1, 8], y=[5, 9], pen="1p,black")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X25c/20c", frame="a")
fig.plot(x=[1, 6, 9], y=[5, 7, 4], pen="1p,black")
fig.show()
```

```python
fig = pygmt.Figure()
fig.basemap(region=[0, 10, 0, 10], projection="X25c/20c", frame="a")
fig.plot(x=[1, 6, 9], y=[5, 7, 4], pen="2p,blue")
fig.plot(x=[2, 4, 10], y=[3, 8, 9], pen="2p,red")
fig.show()
```

```python
fig = pygmt.Figure()
fig.plot(region=[0, 10, 0, 10], projection="X25c/20c", x=[1, 8], y=[5, 9], pen="1p,black")
fig.plot(x=[2, 4, 10], y=[3, 8, 9], pen="2p,red")
fig.show()
```

```python
fig = pygmt.Figure()
fig.plot(
    region=[0, 10, 0, 10],
    projection="X25c/20c",
    frame="a",
    x=[1, 8],
    y=[3, 9],
    pen="5p",
)
fig.show()
```
