```python
import pygmt
```

```python

fig = pygmt.Figure()
fig.coast(
        region="MG+r2", land="brown", water="lightblue", shorelines="thin", frame="a"
    )
with fig.inset(location="jTL+w3.5c+o0.2c", margin=0, border="+pgreen", V=True):
    fig.coast(
    region="g", projection="G47/-20/4c", land="gray", water="white", E="MG+gred"
        )
fig.show()
```
