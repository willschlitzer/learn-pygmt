```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.coast(
    R="-30/30/-40/40",
    J="M25c",
    B="afg",
    I="1/1p,black",
    N="1/0.5p,-",
    W="0.25p,white",
    G="moccasin",
    S="skyblue",
    D="i",
    A=1000,
    L="jCM+c1+w1000k+f+l",
    U=True,
    X="a4c",
    Y="a10c",
    p="135/25",
    t=13,
    #E="MA+gred",
    C="blue"
    )
fig.show()
```

```python
fig = pygmt.Figure()
fig.coast(
    #region="-30/30/-40/40", #R
    region=[-30,30,-40,40], #R
    projection="M25c", #J
    frame="afg", #B
    rivers="1/1p,black", #I
    borders="1/0.5p,-", #N
    shorelines="0.25p,white", #W
    land="moccasin", #G
    water="skyblue", #S
    resolution="i", #D
    area_thresh=1000, #A
    map_scale="jMM+c1+w1000k+f+l", #L
    timestamp=True, #U
    verbose="q", #V
    xshift = "a4c", #X
    yshift = "a10c", #Y
    perspective = "135/25",#p
    transparency = 13, #t
    #dcw="MA+gred",
    lakes="blue"
)
fig.show()
```
