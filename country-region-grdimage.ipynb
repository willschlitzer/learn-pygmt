```python
import pygmt
grid = pygmt.datasets.load_earth_relief(resolution="05m", region="MG+r2")
```

```python
fig=pygmt.Figure()
```

```python
#fig.grdimage(grid="@earth_relief_05m", region="MG+r2")
```

```python
fig.grdimage(grid=grid, cmap="geo", projection="M6i")
fig.coast(shorelines="thin")
fig.basemap(frame=["af","WSne"])
fig.coast(region="g", projection="G47/-20/2i", land="grey", water="white", xshift="1", yshift="16", frame=False)
```

```python
fig.show()
```
