```python
import pygmt
grid = pygmt.datasets.load_earth_relief(resolution="15s", region=[-92, -90, -1.5, 0.5])
```

```python
fig = pygmt.Figure()
fig.grdcontour(grid=grid, projection="M4i")
fig.basemap(frame="x")
fig.basemap(frame="y")
#fig.show()
```

```python
fig = pygmt.Figure()
fig.grdcontour(grid=grid, projection="M4i")
fig.basemap(frame=["xaf", "yg"])
#fig.basemap(frame="xafg")
fig.show()
```
