```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.coast(region="d", projection="W0/15c", land="darkgreen", water="lightblue")
fig.solar(Td="+d1990-02-17T04:27:00")
fig.show()
```

```python
fig = pygmt.Figure()
fig.solar(region="d", projection="W0/15c", Td="+d1990-02-17T04:25:00", pen="1p,red,-.-")
fig.show()
```
