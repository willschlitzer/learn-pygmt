```python
import pygmt
```

Creates initial figure for the PyGMT figure

```python
fig = pygmt.Figure()
```

```python
fig.coast(
    region="g",
    projection="R20/8i",
    shorelines=True,
    water="lightblue",
    land="grey",
    frame=True,
)
```

```python
fig.show()
#fig.show(method="external")
```

```python
ak_fig = pygmt.Figure()
```

```python
ak_fig.coast(
    region=[171, 231, 50, 72],
    projection="S201/90/8i",
    shorelines=True,
    water="lightblue",
    land="grey",
    frame="a10f2",
    resolution="i"
)
```

```python
ak_fig.show()
```

```python
from numpy.random import randint

POINT_FILL = "red"
NUM_PTS = 5

lons = randint(-160, -141, NUM_PTS)
lats = randint(55, 71, NUM_PTS)

ak_fig.plot(
    x=lons,
    y=lats,
    style="a0.2i",
    color=POINT_FILL,
    pen="black",
    label=f"{POINT_FILL}",
)
#ak_fig.legend()
```

```python
ak_fig.show()
```
