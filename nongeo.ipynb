```python
import pygmt
```

```python
fig = pygmt.Figure()
# ``region`` sets the x and y ranges or the Cartesian figure.
# The argument ``WSne`` is passed to ``frame`` to put axis labels only on the left and bottom axes.
# The ``x`` and ``y`` parameters are used to plot lines on the figure.
fig.plot(
    x=[3, 9, 2],
    y=[4, 9, 37],
    pen="3p,red",
    region=[0, 10, 0, 50],
    projection="X15c/10c",
    frame=["af", "WSne"],
)
fig.show()
```

```python


fig = pygmt.Figure()
fig.plot(
    # x inputs are the theta values for a polar plot
    x=[180, 120, 270, 60, 0],
    # y inputs are the radius values for a polar plot
    y=[15, 35, 15, 35, 15],
    pen="2p,blue",
    # the region values are theta-min/theta-max/radius-min/radius-max
    region=[0, 360, 0, 40],
    projection="P15c",
    frame=["afg"],
)
fig.show()
```

```python
import numpy as np
xline = np.arange(0, 101)
yline = xline ** .5
xpoints = np.arange(0, 101, 10)
ypoints = xpoints ** .5

fig = pygmt.Figure()
fig.plot(region=[1, 100, 0, 10], projection="X15cl/10c",
         frame=["WSne+gbisque", "x2g3", "ya2f1g2"], x=xline, y=yline,
        pen="1p,blue,-")
fig.plot(x=xpoints, y=ypoints, style="s0.3c", color="red", no_clip="c", pen="black")
fig.show()
```

```python
import numpy as np
yvalues = np.arange(0, 11)
xvalues= yvalues ** 2

fig = pygmt.Figure()
fig.plot(region=[0, 100, 0, 10], projection="X15p0.5/10c",
         frame=["WSne+givory", "xa1p", "ya2f1"], pen="thick",
        x=xvalues, y=yvalues)
fig.plot(x=xvalues, y=yvalues, style="c0.2c", color="green", no_clip="c", pen="black")
fig.show()
```
