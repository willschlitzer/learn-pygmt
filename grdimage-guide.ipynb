```python
import pygmt
```

```python
fig = pygmt.Figure()
fig.grdimage(grid="@earth_relief_30m")
fig.show()
```

```python
fig = pygmt.Figure()
fig.grdimage(grid="@earth_relief_30m", projection="R0/5i")
fig.show()
```

```python
# Orthographic
fig = pygmt.Figure()
fig.grdimage(grid="@earth_relief_30m", projection="G10/52/4.5i")
fig.show()
```

```python
# Azimuthal Equidistant
fig = pygmt.Figure()
fig.grdimage(grid="@earth_relief_30m", projection="E-100/40/4.5i")
fig.show()
```

```python
# Mercator
fig = pygmt.Figure()
fig.grdimage(grid="@earth_relief_30m", projection="M5i", region=[-180, 180, -75, 75])
fig.show()
```
