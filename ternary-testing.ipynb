```python
data_list = []
for _ in range(30):
    a = np.random.randint(0, 100)
    b = np.random.randint(0, 100-a)
    c = 100-(a+b)
    value = round(np.random.randint(0, 4)+np.random.random(), 3)
    data_list.append([a/100, b/100, c/100, value])
print(data_list)
```

```python
fig = pygmt.Figure()
fig.ternary(data=np.array(data_list),
            region=[0, 100, 0, 100, 0, 100],
            cmap="red,orange,yellow,green,blue,violet",
            width="6i",
            frame=["bafg+lAir", "cafg+lLimestone", "aafg+lWater"],
            style="c0.1c",
            pen="thinnest",
           )
fig.show()
```
